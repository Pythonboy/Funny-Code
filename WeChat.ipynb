```python
import itchat
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
itchat.auto_login(hotReload = True);
```

```python
'''统计微信中的男女性别'''
friends = itchat.get_friends(update=True)[:]
total = len(friends) - 1
man = women = other = 0
for friend in friends[0:] :
    sex = friend["Sex"]
    if sex == 1 :
        man += 1
    elif sex == 2 :
        women += 1
    else :
        other += 1  
print("男性好友：%.2f%%" % (float(man) / total * 100))
print("女性好友：%.2f%%" % (float(women) / total * 100))
print("其他：%.2f%%" % (float(other) / total * 100))

```

```python
plt.figure();
Class = ["men","women","others"];
number = [float(man)/total*100,float(women)/total*100,float(other)/total*100];
plt.pie(number,labels = Class,autopct = "%.2f%%");
plt.title("微信中男女比例",fontproperties = 'SimHei',fontsize = 20)
plt.legend()
plt.axis('equal')
plt.show()
```
